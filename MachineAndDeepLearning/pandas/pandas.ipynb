# Reading data in pandas

```python
df=raw_data.copy()
```

```python
#Read all the columns present in the data frame
display(df.columns)
```

```python
#Displa all the names of the pokemon in the table
display(df['Name'])
```

```python
#Read multiple columns in one command
display(df[['Name','Type 1','HP']])
```

```python
## iloc[start:end rows], used to read the data from a specific location
display(df.iloc[0:4])
```

```python
#we can use iloc to read data from a specific location
print(df.iloc[2,0:12])
```

```python
#Iterating in the table
for index, row in df.iterrows():
    print(index,row)
```

```python
#loc()->Display some rows with certain condition , and operation will not work in this
df.loc[df['HP']>45]
```

```python
# describe() -> FOr describing the data
df.describe()
```

```python
# Arranging data into Ascending and Decending order
df.sort_values('Name', ascending=False)
```
