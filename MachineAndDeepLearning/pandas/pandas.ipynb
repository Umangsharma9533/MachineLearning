# Reading data in pandas

```python
df=raw_data.copy()
```

```python
#Read all the columns present in the data frame
display(df.columns)
```

```python
#Displa all the names of the pokemon in the table
display(df['Name'])
```

```python
#Read multiple columns in one command
display(df[['Name','Type 1','HP']])
```

```python
## iloc[start:end rows], used to read the data from a specific location
display(df.iloc[0:4])
```

```python
#we can use iloc to read data from a specific location
print(df.iloc[2,0:12])
```

```python
#Iterating in the table
for index, row in df.iterrows():
    print(index,row)
```

```python
#loc()->Display some rows with certain condition , and operation will not work in this
df.loc[df['HP']>45]
```

```python
# describe() -> FOr describing the data
df.describe()
```

```python
# Arranging data into Ascending and Decending order
df.sort_values('Name', ascending=False)
```

```python
df.to_csv('modified.csv', index=False)

df.to_excel('modified.xlsx', index=False)

df.to_csv('modified.txt', index=False, sep='\t')
```

```python
#FIltering the data
df.loc[(df['Type 1']=='Grass')| (df['Type 2']=='Poison')]
```

```python
df.loc[(df['Type 1']=='Grass')| (df['Type 2']=='Poison') & (df['HP']>40)]
```

```python
import re #Importing Regular Expression
# re.I -Ignore the Lower and Upper Case in regular expressions
#regex = True,Means it will follow regular expression
df.loc[df['Type 1'].str.contains('grass|fire', flags=re.I,regex=True)]
```

```python
# Display all the name starts with pi
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I,regex=True)]
```

```python
#If we want to rename and particular value in the table use
#df.loc[df['Type 1']=='Fire', 'Type 1']='Flammer'
#display(df)
```

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

#Reset the value of index to make further processiing of data easy
new_df.reset_index(drop=True, inplace=True)

new_df

new_df.to_csv('filtered.csv')
```
